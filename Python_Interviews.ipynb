```python
# 49. How do you find the factorial of a number using recursion?

''''
* Recursion is the concept where a function can call itself any number of times till a certain condition is met.
* The condition is important to avoid infinite looops
* Let's check this out practically!

'''''

def factorial(a):
    if a == 1:
        return 1
    else:
        return (a * factorial(a-1))

input_num = 3
print("The factorial of", input_num, "is", factorial(input_num))
```
